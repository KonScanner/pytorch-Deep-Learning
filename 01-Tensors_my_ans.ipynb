## Python native data types

Python has many native datatypes. Here are the important ones:

 - **Booleans** are either `True` or `False`.
 - **Numbers** can be integers (1 and 2), floats (1.1 and 1.2), fractions (1/2 and 2/3), or even complex numbers.
 - **Strings** are sequences of Unicode characters, e.g. an html document.
 - **Lists** are ordered sequences of values.
 - **Tuples** are ordered, immutable sequences of values.
 - **Sets** are unordered bags of values.
 - **Dictionaries** are unordered bags of key-value pairs.
 
See [here](http://www.diveintopython3.net/native-datatypes.html) for a complete overview.

### Torch!

```python
# Generate a tensor of size 3x4x5
t = torch.Tensor(3, 4, 5)
type(t)
```

```python
# t.size() is a classic tuple
print(f"{t.size()=}")
```

```python
# prints dimensional space and sub-dimensions
print(f"Point in a {t.numel()=} dimensional space")
print(f"Organised in {t.dim()=} sub dimensions")
```

```python
t
```

```python
# Mind the underscore!
# Any operation that mutates a tensor in-place is post-fixed with an _.
# For example: x.copy_(y), x.t_(), x.random_(n) will mutate x
t.random_(10)
```

```python
# This mutates the tensor permanently
r = torch.Tensor(t)
r.resize_(3,8)
```

```python
# As you can see zero_ would replace r with 0's which was originally filled with integers
r.zero_()
```

```python
t
```

```python
# This is* important
s = r.clone()
```

```python
s
```

```python
# In-place fill of 42's
s.fill_(42)
```

```python
# Because r was cloned, even though we did an in place operation, this still does not affect r
r
```

## Vectors (1D Tensors)

```python
# Creates a 1D Tensor of ints {4,3,2,1}
v = torch.Tensor([4, 3, 2, 1])
v
```

```python
print(f"{v.dim()}, {v.size()[0]}")
```

```python
w = torch.Tensor([1, 0, 2, 0])
w
```

```python
# Element wise multiplication
v*w
```

```python
# Scalar Product 4*1 + 3*0 + 2*2 + 1*0
v@w
```

```python
# In-place replacement of random number from 0 to 10
x = torch.Tensor(5).random_(10)
x
```

```python
# Extract sub-tensor [from:to)
x[1:2 +1]
```

## Matrices (2D Tensors)

```python
# Create a 2x4 Tensor
m = torch.Tensor([[5, 3, 2, 7], [4, 9, 2, 1]])
m
```

```python
print(f"{m.dim()=}")
```

```python
print(f"{m.size(0)=}", f"{m.size(1)=}", f"{m.size()=}", sep=', ')
```

```python
# Return the total number of elements in the tensor
print(f"{m.numel()=}")
```

```python
# Indexing row 0, column 2 (0-indexed)
m[0][2]
```

```python
# Indexing row 0, column 2 (0-indexed)
m[0, 2]
```

```python
# Indexing column 1, all rows (returns size 2)
m[:, 1]
```

```python
# Indexing column 1, all rows (returns size 2x2)
m[:, [1]]
```

```python
# Indexing row 0, all columns (returns size 1x4)
m[[0], :]
```

```python
# Indexing row 0, all columns (returns size 1x4)
m[[0], :]
```

```python
# Indexes row 0, all columns (returns size 1x4)
m[0, :]
```

```python
# Create Tensor of numbers from 1 to 5 (exluding 5)
v = torch.arange(1., 4+1)
v
```

```python
m
```

```python
m@v # Scalar Product
```

```python
# Calculated by 4.*1+ 9.*2 +2.*3 + 1*1.
m[[1], :]@v
```

```python
# Calculated by 5.*1+ 3.*2 +2.*3 + 7.*4.
m[[0], :]@v
```

```python
# Add a random tensor of size 2x4 to m
m + torch.rand(2, 4)
```

```python
# Subtract a random tensor of size 2x4 to m
m - torch.rand(2, 4)
```

```python
# Multiply a random tensor of size 2x4 to m
m * torch.rand(2, 4)
```

```python
# Divide m by a random tensor of size 2x4
m / torch.rand(2, 4)
```

```python
m.t() # Transpose of m same as m.transpose(0, 1)
```

```python
m.transpose(0, 1) # Transpose of m same as m.t()
```

## Constructors

```python
# Create a tensor of integers ranging from 0 to 5
import numpy as np
n_np = np.arange(0, 6)
n = torch.from_numpy(n_np) # Not a clone, points to same memory
print(n_np,n)
```

```python
# In-place multiplication of all elements by 2 for n
# Because n is essnetial n_np, not a clone, this affects n_np
n.mul_(2)
n_np
```
